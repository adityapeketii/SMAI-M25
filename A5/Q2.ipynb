```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.preprocessing import StandardScaler
import pickle

import os
output_directory = "Outputs/Q2_new"
os.makedirs(output_directory, exist_ok=True)


class TimeSeriesDataset(Dataset):
    def __init__(self, data, history_length):
        self.data = data
        self.history_length = history_length
        
    def __len__(self):
        return len(self.data) - self.history_length
    
    def __getitem__(self, idx):
        x = self.data[idx:idx+self.history_length]
        y = self.data[idx+self.history_length]
        return torch.FloatTensor(x), torch.FloatTensor([y])

class SimpleRNN(nn.Module):
    def __init__(self, input_size=1, hidden_size=32, num_layers=2, dropout=0.2):
        super(SimpleRNN, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.rnn = nn.RNN(input_size, hidden_size, num_layers, 
                         batch_first=True, dropout=dropout if num_layers > 1 else 0)
        self.fc = nn.Linear(hidden_size, 1)
        
    def forward(self, x):
        out, _ = self.rnn(x)
        out = self.fc(out[:, -1, :])
        return out

class LinearAR(nn.Module):
    def __init__(self, history_length):
        super(LinearAR, self).__init__()
        self.fc = nn.Linear(history_length, 1)
        
    def forward(self, x):
        x = x.squeeze(-1)
        return self.fc(x)

class MLPPredictor(nn.Module):
    def __init__(self, history_length, hidden_size=64):
        super(MLPPredictor, self).__init__()
        self.net = nn.Sequential(
            nn.Linear(history_length, hidden_size),
            nn.ReLU(),
            nn.Linear(hidden_size, hidden_size//2),
            nn.ReLU(),
            nn.Linear(hidden_size//2, 1)
        )
        
    def forward(self, x):
        x = x.squeeze(-1)
        return self.net(x)

def load_and_split_data(filepath, train_frac=0.7, val_frac=0.15):
    df = pd.read_csv(filepath)
    data = df.iloc[:, 0].values
    
    n = len(data)
    train_end = int(n * train_frac)
    val_end = int(n * (train_frac + val_frac))
    
    train_data = data[:train_end]
    val_data = data[train_end:val_end]
    test_data = data[val_end:]
    
    scaler = StandardScaler()
    train_data_norm = scaler.fit_transform(train_data.reshape(-1, 1)).flatten()
    val_data_norm = scaler.transform(val_data.reshape(-1, 1)).flatten()
    test_data_norm = scaler.transform(test_data.reshape(-1, 1)).flatten()
    
    return (train_data_norm, val_data_norm, test_data_norm), scaler, (train_data, val_data, test_data)


def train_model(model, train_loader, val_loader, epochs=50, lr=0.001, device='cpu'):
    model = model.to(device)
    criterion = nn.MSELoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    
    train_losses = []
    val_losses = []
    
    for epoch in range(epochs):
        model.train()
        train_loss = 0
        for x, y in train_loader:
            x, y = x.to(device), y.to(device)
            if isinstance(model, SimpleRNN):
                x = x.unsqueeze(-1)
            
            optimizer.zero_grad()
            outputs = model(x)
            loss = criterion(outputs, y)
            loss.backward()
            optimizer.step()
            train_loss += loss.item()
        
        model.eval()
        val_loss = 0
        with torch.no_grad():
            for x, y in val_loader:
                x, y = x.to(device), y.to(device)
                if isinstance(model, SimpleRNN):
                    x = x.unsqueeze(-1)
                outputs = model(x)
                loss = criterion(outputs, y)
                val_loss += loss.item()
        
        train_losses.append(train_loss / len(train_loader))
        val_losses.append(val_loss / len(val_loader))
        
        if (epoch + 1) % 10 == 0:
            print(f'Epoch [{epoch+1}/{epochs}], Train Loss: {train_losses[-1]:.6f}, Val Loss: {val_losses[-1]:.6f}')
    
    return train_losses, val_losses

def predict_single_step(model, data, history_length, device='cpu'):
    model.eval()
    predictions = []
    
    with torch.no_grad():
        for i in range(len(data) - history_length):
            x = torch.FloatTensor(data[i:i+history_length]).unsqueeze(0).to(device)
            if isinstance(model, SimpleRNN):
                x = x.unsqueeze(-1)
            pred = model(x)
            predictions.append(pred.cpu().item())
    
    return np.array(predictions)

def autoregressive_forecast(model, initial_history, forecast_length, device='cpu'):
    model.eval()
    history = list(initial_history)
    predictions = []
    
    with torch.no_grad():
        for _ in range(forecast_length):
            x = torch.FloatTensor(history[-len(initial_history):]).unsqueeze(0).to(device)
            if isinstance(model, SimpleRNN):
                x = x.unsqueeze(-1)
            pred = model(x)
            pred_val = pred.cpu().item()
            predictions.append(pred_val)
            history.append(pred_val)
    
    return np.array(predictions)

def fit_analytical_recurrence(data, order=2):
    X = []
    y = []
    
    for i in range(order, len(data)):
        X.append(data[i-order:i][::-1])
        y.append(data[i])
    
    X = np.array(X)
    y = np.array(y)
    
    coeffs = np.linalg.lstsq(X, y, rcond=None)[0]
    
    return coeffs

def analytical_predict(coeffs, data, order):
    predictions = []
    
    for i in range(order, len(data)):
        history = data[i-order:i][::-1]
        pred = np.dot(coeffs, history)
        predictions.append(pred)
    
    return np.array(predictions)

def analytical_autoregressive(coeffs, initial_history, forecast_length):
    history = list(initial_history)
    predictions = []
    order = len(coeffs)
    
    for _ in range(forecast_length):
        recent = np.array(history[-order:])[::-1]
        pred = np.dot(coeffs, recent)
        predictions.append(pred)
        history.append(pred)
    
    return np.array(predictions)

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

path = "Dataset/Q2/recurrence_timeseries.csv"
(train_norm, val_norm, test_norm), scaler, (train_raw, val_raw, test_raw) = load_and_split_data(path)

print(f"Train size: {len(train_norm)}, Val size: {len(val_norm)}, Test size: {len(test_norm)}")

history_lengths = [2, 3, 5, 10]
best_history = None
best_val_loss = float('inf')

for p in history_lengths:
    train_dataset = TimeSeriesDataset(train_norm, p)
    val_dataset = TimeSeriesDataset(val_norm, p)
    
    train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=64, shuffle=False)
    
    model = SimpleRNN(input_size=1, hidden_size=32, num_layers=2)
    train_losses, val_losses = train_model(model, train_loader, val_loader, epochs=30, device=device)
    
    final_val_loss = val_losses[-1]
    print(f"History length {p}: Final validation loss = {final_val_loss:.6f}")
    
    if final_val_loss < best_val_loss:
        best_val_loss = final_val_loss
        best_history = p

print(f"\nBest history length: {best_history}")

p = best_history
train_dataset = TimeSeriesDataset(train_norm, p)
val_dataset = TimeSeriesDataset(val_norm, p)
test_dataset = TimeSeriesDataset(test_norm, p)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=64, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)

models = {
    'LinearAR': LinearAR(p),
    'MLP': MLPPredictor(p, hidden_size=64),
    'RNN': SimpleRNN(input_size=1, hidden_size=32, num_layers=2)
}

trained_models = {}
results = {}

for name, model in models.items():
    print(f"\nTraining {name}...")
    train_losses, val_losses = train_model(model, train_loader, val_loader, epochs=50, device=device)
    trained_models[name] = model
    results[name] = {'train_losses': train_losses, 'val_losses': val_losses}
    
    plt.figure(figsize=(10, 5))
    plt.plot(train_losses, label='Train Loss')
    plt.plot(val_losses, label='Val Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title(f'{name} Training History')
    plt.legend()
    plt.grid(True)
    plt.text(0.95, 0.95, "peketivenkata.v", ha='right', va='top',
             transform=plt.gca().transAxes, fontsize=10, color='gray', alpha=0.7)
    
    plt.savefig(os.path.join(output_directory, f"train_curve_{name}.png"), dpi=300, bbox_inches='tight')
    plt.show()


test_predictions = {}
for name, model in trained_models.items():
    preds = predict_single_step(model, test_norm, p, device)
    test_predictions[name] = preds

print("\n=== Single-Step Prediction Results ===")
for name, preds in test_predictions.items():
    actual = test_norm[p:]
    mae = np.mean(np.abs(preds - actual))
    mse = np.mean((preds - actual)**2)
    print(f"{name} - MAE: {mae:.6f}, MSE: {mse:.6f}")

coeffs = fit_analytical_recurrence(train_norm, order=p)
print(f"\nAnalytical Recurrence Coefficients (order {p}): {coeffs}")
print(f"Recurrence: x_k = {' + '.join([f'{c:.4f}*x_(k-{i+1})' for i, c in enumerate(coeffs)])}")

analytical_preds = analytical_predict(coeffs, test_norm, p)
actual = test_norm[p:]
mae_analytical = np.mean(np.abs(analytical_preds - actual))
mse_analytical = np.mean((analytical_preds - actual)**2)
print(f"Analytical - MAE: {mae_analytical:.6f}, MSE: {mse_analytical:.6f}")

forecast_horizons = [1, 5, 10, 20, 50, 100]
forecast_results = {name: [] for name in trained_models.keys()}
forecast_results['Analytical'] = []

initial_history = test_norm[:p]

for horizon in forecast_horizons:
    for name, model in trained_models.items():
        ar_preds = autoregressive_forecast(model, initial_history, horizon, device)
        actual_future = test_norm[p:p+horizon]
        mae = np.mean(np.abs(ar_preds - actual_future))
        forecast_results[name].append(mae)
    
    analytical_ar_preds = analytical_autoregressive(coeffs, initial_history, horizon)
    actual_future = test_norm[p:p+horizon]
    mae = np.mean(np.abs(analytical_ar_preds - actual_future))
    forecast_results['Analytical'].append(mae)

print("\n=== Autoregressive Forecasting Results ===")
plt.figure(figsize=(12, 6))
for name, errors in forecast_results.items():
    plt.plot(forecast_horizons, errors, marker='o', label=name)
plt.xlabel('Forecast Horizon')
plt.ylabel('MAE')
plt.title('Forecast Error vs Horizon Length')
plt.legend()
plt.grid(True)
plt.text(0.95, 0.95, "peketivenkata.v", ha='right', va='top',
         transform=plt.gca().transAxes, fontsize=10, color='gray', alpha=0.7)
plt.savefig(os.path.join(output_directory, "forecast_error_vs_horizon.png"), dpi=300, bbox_inches='tight')
plt.show()


param_counts = {}
for name, model in trained_models.items():
    param_counts[name] = sum(p.numel() for p in model.parameters())
    print(f"{name} parameters: {param_counts[name]}")

print(f"Analytical recurrence parameters: {p}")

plt.figure(figsize=(14, 6))
plt.plot(actual[:500], label='Actual', alpha=0.7)
plt.plot(test_predictions['RNN'][:500], label='RNN Prediction', alpha=0.7)
plt.plot(analytical_preds[:500], label='Analytical Prediction', alpha=0.7)
plt.xlabel('Time Step')
plt.ylabel('Value')
plt.title('Predictions vs Actual (First 500 test points)')
plt.legend()
plt.grid(True)
plt.text(0.95, 0.95, "peketivenkata.v", ha='right', va='top',
         transform=plt.gca().transAxes, fontsize=10, color='gray', alpha=0.7)
plt.savefig(os.path.join(output_directory, "prediction_vs_actual_first500.png"), dpi=300, bbox_inches='tight')
plt.show()

# === 1. Actual vs RNN ===
plt.figure(figsize=(14, 6))
plt.plot(actual[:500], label='Actual', alpha=0.8)
plt.plot(test_predictions['RNN'][:500], label='RNN', alpha=0.8)
plt.xlabel('Time Step')
plt.ylabel('Value')
plt.title('Actual vs RNN Prediction (First 500 Points)')
plt.legend()
plt.grid(True)
plt.text(0.95, 0.95, "peketivenkata.v", ha='right', va='top',
         transform=plt.gca().transAxes, fontsize=10, color='gray', alpha=0.7)
plt.savefig(os.path.join(output_directory, "actual_vs_rnn.png"), dpi=300, bbox_inches='tight')
plt.show()


# === 2. Actual vs MLP ===
plt.figure(figsize=(14, 6))
plt.plot(actual[:500], label='Actual', alpha=0.8)
plt.plot(test_predictions['MLP'][:500], label='MLP', alpha=0.8)
plt.xlabel('Time Step')
plt.ylabel('Value')
plt.title('Actual vs MLP Prediction (First 500 Points)')
plt.legend()
plt.grid(True)
plt.text(0.95, 0.95, "peketivenkata.v", ha='right', va='top',
         transform=plt.gca().transAxes, fontsize=10, color='gray', alpha=0.7)
plt.savefig(os.path.join(output_directory, "actual_vs_mlp.png"), dpi=300, bbox_inches='tight')
plt.show()


# === 3. Actual vs LinearAR ===
plt.figure(figsize=(14, 6))
plt.plot(actual[:500], label='Actual', alpha=0.8)
plt.plot(test_predictions['LinearAR'][:500], label='Linear AR', alpha=0.8)
plt.xlabel('Time Step')
plt.ylabel('Value')
plt.title('Actual vs Linear AR Prediction (First 500 Points)')
plt.legend()
plt.grid(True)
plt.text(0.95, 0.95, "peketivenkata.v", ha='right', va='top',
         transform=plt.gca().transAxes, fontsize=10, color='gray', alpha=0.7)
plt.savefig(os.path.join(output_directory, "actual_vs_linearAR.png"), dpi=300, bbox_inches='tight')
plt.show()


# === 4. Actual vs Analytical Recurrence ===
plt.figure(figsize=(14, 6))
plt.plot(actual[:500], label='Actual', alpha=0.8)
plt.plot(analytical_preds[:500], label='Analytical', alpha=0.8)
plt.xlabel('Time Step')
plt.ylabel('Value')
plt.title('Actual vs Analytical Prediction (First 500 Points)')
plt.legend()
plt.grid(True)
plt.text(0.95, 0.95, "peketivenkata.v", ha='right', va='top',
         transform=plt.gca().transAxes, fontsize=10, color='gray', alpha=0.7)
plt.savefig(os.path.join(output_directory, "actual_vs_analytical.png"), dpi=300, bbox_inches='tight')
plt.show()



best_model_name = None
best_model_val = float('inf')

for name, hist in results.items():
    val_last = hist['val_losses'][-1]
    if val_last < best_model_val:
        best_model_val = val_last
        best_model_name = name

print("\n=== Analysis Summary ===")
print(f"Best model based on validation loss: {best_model_name} with {param_counts[best_model_name]} parameters")
```

Analysis Summary

- Analytical model has only 10 parameters but achieves competitive performance and even better.

- This suggests the underlying dynamics can be captured by a simple linear recurrence.

- The data appears to follow an order-10 autoregressive process.

- Temporal relation: xk-depends linearly on the previous 10 values.
